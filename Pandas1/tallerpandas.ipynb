## Punto 3

```python
#Genermos copia de df ya que se van a operar los datos
df2 = df.copy()
```

```python
# Convertir la colummna fechainicio y fechafin al tipo de dato datetime
df2['FechaInicio'] = pd.to_datetime(df2['FechaInicio'])
df2['FechaFin'] = pd.to_datetime(df2['FechaFin'])
```

```python
#Dividir las columnas de fechas en ano y mes
df2['Ano_Inicio'] = df2['FechaInicio'].dt.year
df2['Mes_Inicio'] = df2['FechaInicio'].dt.month
df2['Dia_Inicio'] = df2['FechaInicio'].dt.day
```

```python
#Dividir las columnas de fechas fin en ano y mes
df2['Ano_Fin'] = df2['FechaFin'].dt.year
df2['Mes_Fin'] = df2['FechaFin'].dt.month
df2['Dia_Fin'] = df2['FechaFin'].dt.day
```

## Punto 4

```python
#identificar valores nulos
valores_nulos = df2.isnull().sum()
valores_nulos
```

## Punto 5

```python
 # Seleccionar solo las columnas numéricas del DataFrame
 df_numericas = df2.select_dtypes(include=[np.number])
 df_numericas
```

```python
# Calcular la matriz de correlación de Pearson (el método por defecto)
correlacion_person = df_numericas.corr()
```

```python
#matriz de correlacion de pearson
figcorrelacion = px.imshow(correlacion_person, text_auto=True)
figcorrelacion.show()
```

```python
# Calcular la matriz de correlación de sperman (el método por defecto)
correlacion_spearman = df_numericas.corr(method='spearman')
```

```python
figcorrelacion_spearman = px.imshow(correlacion_spearman, text_auto=True)
figcorrelacion_spearman.show()
```

```python
#Cantidad de aportes por mes y año
df2.groupby(['Ano_Inicio','Mes_Inicio']).agg({'Etapa': 'count', 'AportesProyectados': 'sum'}).reset_index()
```

```python
#Cantidad de aportes por mes y año
aportes_mes = df2.groupby(['Ano_Inicio','Mes_Inicio']).agg({'Etapa': 'count', 'AportesProyectados': 'sum'}).reset_index()
aportes_mes.rename(columns={'Etapa': 'Cantidad Registros'}, inplace=True)
```

```python
# Ordenar los resultados de acuerdo al año y mes
aportes_mes = aportes_mes.sort_values(by=['Ano_Inicio','Mes_Inicio'])
aportes_mes
```

```python
# Mes de cada año donde se realizan más aportes
max = aportes_mes.groupby('Ano_Inicio')['AportesProyectados'].idxmax()
```

```python
mes_mas_costoso = aportes_mes.loc[max]
mes_mas_costoso
```

```python
min =aportes_mes.groupby('Ano_Inicio')['AportesProyectados'].idxmin()
```

```python
mes_menos_costoso = aportes_mes.loc[min]
```

```python
#Contar registros por etapa
conteo_por_etapa = df2.groupby('Etapa')['Etapa'].count().reset_index(name='Cantidad')
conteo_por_etapa
```

```python
#Graficar
fig = px.pie(conteo_por_etapa, values='Cantidad', names='Etapa', title='Cantidad de Registros por Etapa')
fig.show()
```
